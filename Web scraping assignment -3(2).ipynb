# Q.2
In the above question, now scrape the following details of each product listed in first 3 pages of your search
results and save it in a data frame and csv. In case if any product has less than 3 pages in search results then
scrape all the products available under that product name. Details to be scraped are: "Brand
Name", "Name of the Product", "Price", "Return/Exchange", "Expected Delivery", "Availability" and
“Product URL”. In case, if any of the details are missing for any of the product then replace it by “-“

```python
Brand =[]
Product = []
Rating =[]
No_of_ratings =[]
Price =[]
Return =[]
Excepted_delivery =[]
Availability =[]
Other_details =[]
```

```python
URL= []
for page in range(0,3):
    url=driver.find_elements(By.XPATH,'//a[@class="a-link-normal a-text-normal"]')
    for i in url:
        URL.append(i.get_attribute('href'))
    time.sleep(2)
```

```python
next_page = driver.find_element(By.XPATH,'//a[@class="s-pagination-item s-pagination-next s-pagination-button s-pagination-separator"]')
next_page.click()
```

```python
from tqdm import tqdm
for i in tqdm(URL):
    driver.get(i)
    time.sleep(2)
    
    try:
        brand=driver.find_element(By.XPATH,'//a[@id="bylineInfo"]')
        Brand.append(brand.text) 
    except NoSuchElementException:
        Brand.append('-')
        
    time.sleep(1)
    
    try:
        product =driver.find_element(By.XPATH,'//span[@id="productTitle"]')
        Product.append(product.text)
    except NoSuchElementException:
        Product.append('-')
    time.sleep(1)
    
    try:
        rating=driver.find_element(By.XPATH,'//span[@class="a-size-medium a-color-base"]')
        Rating.append(rating.text)
    except NoSuchElementException:
        Rating.append('-')
    time.sleep(1)

    try:
        rating_count=driver.find_element(By.XPATH,'//span[@class="a-size-base a-color-secondary"]')
        No_of_ratings.append(rating_count.text)
    except NoSuchElementException:
        No_of_ratings.append('-')
    time.sleep(1)
    
    try:
        price=driver.find_element(By.XPATH,'//span[@id="priceblock_dealprice"]')
        Price.append(price.text)
    except NoSuchElementException:
        Price.append('-')
    time.sleep(1)
    
    try:
        replacement=driver.find_element(By.XPATH,'//*[@id="RETURNS_POLICY"]/span/div[2]/a')
        Return.append(replacement.text)
    except NoSuchElementException:
        Return.append('-')
    time.sleep(1)
    
    try:
        delivery=driver.find_element(By.XPATH,'//div[@id="ddmDeliveryMessage"]/b')
        Excepted_delivery.append(delivery.text)
    except NoSuchElementException:
        Excepted_delivery.append('-')
    time.sleep(1)
    
    try:
        availability=driver.find_element(By.XPATH,'//span[@class="a-size-medium a-color-success"]')
        Availability.append(availability.text)
    except NoSuchElementException:
        Availability.append('-')
    time.sleep(1)
    
    try:
        other=driver.find_element(By.XPATH,'//ul[@class="a-unordered-list a-vertical a-spacing-mini"]')
        Other_details.append(other.text)
    except NoSuchElementException:
        Other_details.append('-')
    time.sleep(1)
```

```python
amazon=pd.DataFrame({})
amazon['Brand']=Brand[:50]
amazon['Product']=Product[:50]
amazon['rating']=Rating[:50]
amazon['No of ratings']=No_of_ratings[:50]
amazon['Price']=Price[:50]
amazon['Return']=Return[:50]
amazon['Excepteddelivery']=Excepted_delivery[:50]
amazon['Price']=Price[:50]
amazon['Availability']=Availability[:50]
amazon
```

# Q.9
Write a python program to scrape a data for all available Hostels from https://www.hostelworld.com/ in
“London” location. You have to scrape hostel name, distance from city centre, ratings, total reviews, overall
reviews, privates from price, dorms from price, facilities and property description.

```python
import selenium
import pandas as pd
import time
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.common.exceptions import StaleElementReferenceException, NoSuchElementException
import requests
from selenium.webdriver.common.by import By
```

```python
driver = webdriver.Chrome()
```

```python
driver.get('https://www.hostelworld.com/')
```

```python
location = driver.find_element(By.XPATH,'//input[@class="native-input"]')
location.click()
```

```python
search = driver.find_element(By.XPATH,'//input[@class="native-input"]')
search.send_keys('London')
```

```python
tap = driver.find_element(By.XPATH,'//div[@class="item-text"]')
tap.click()
```

```python
summit = driver.find_element(By.XPATH,'//button[@class="btn-content medium-button icon-only"]')
summit.click()
```

```python
hostel_name = []
location = []
ratings = []
total_reviews = []
overall_reviews = []
private_price = []
Facilities = []
dorm_price = []
description = []
```

```python
hostel = driver.find_elements(By.XPATH, '//div[@class="property-name"]')
try:
    for i in hostel:
        hostel_name.append(i.text)
except:
    hostel_name.append("-")
```

```python
distance = driver.find_elements(By.XPATH, '//div[@class="property-distance"]')
try:
    for i in distance:
        location.append(i.text)
except:
    location.append("-")
```

```python
information = driver.find_elements(By.XPATH, '//span[@class="number"]')
try:
    for i in information:
        ratings.append(i.text.split("\n")[0])
except:
    ratings.append("-")
```

```python
reviews = driver.find_elements(By.XPATH,'//span[@class="left-margin"]')
try:
    for i in reviews:
        total_reviews.append(i.text)
except:
    total_reviews.append("-")
```

```python
review = driver.find_elements(By.XPATH,'//span[@class="keyword"]')
try:
    for i in review:
        overall_reviews.append(i.text)
except:
    overall_reviews.append("-")
```

```python
private = driver.find_elements(By.XPATH,'//div[@class="accommodation-price unavailable"]')
try:
    for i in private:
        private_price.append(i.text.replace("\n", ""))
except:
    private_price.append("-")
```

```python
dorm = driver.find_elements(By.XPATH,'//strong[@class="current"]')
for i in dorm:
    dorm_price.append(i.text.replace("\n", ""))
```

```python
facilities = driver.find_elements(By.XPATH,'//div[@class="icon-container badge-icon"]')
try:
    for i in facilities:
        Facilities.append(i.text)
except:
    Facilities.append("-")
```

```python
try:
    disc = driver.find_element(By.XPATH,'//div[@class="icon-container badge-icon"]')
    description.append(disc.text)
except NoSuchElementException:
    description.append('-')
```

```python
HOSTEL=pd.DataFrame({})
HOSTEL['Hostel_name']=hostel_name[:20]
HOSTEL['Locations']=location[:20]
HOSTEL['Ratings']=ratings[:20]
HOSTEL['Total_reviews']=total_reviews[:20]
HOSTEL['Overall_reviews']=overall_reviews[:20]
HOSTEL['Private_price']=private_price[:20]
HOSTEL['Dorm_price']=dorm_price[:20]
HOSTEL['facilities']=Facilities[:20]
HOSTEL
```
